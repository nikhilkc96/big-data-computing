```python
try:
    sc.stop()
except:
    pass
from pyspark import SparkContext, SparkConf
from pyspark.sql import SparkSession
sc=SparkContext()
spark = SparkSession(sparkContext=sc)

rdd = sc.textFile("/Users/nikhilkc/Desktop/BDC/data.txt")
rdd.take(5)
```

```python
def Func(lines):
    lines =lines.lower()
    lines =lines.split()
    return lines
rdd1 =rdd.map(Func)
print(rdd1.take(5))
rddflat  =rdd.flatMap(Func)
stopwords = rddflat.take(5)
```

```python
rdd3 =rddflat.filter(lambda x: x not in stopwords)
rdd3.take(10)
```
